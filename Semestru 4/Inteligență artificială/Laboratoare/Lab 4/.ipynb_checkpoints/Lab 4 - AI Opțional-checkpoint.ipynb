```python
import random
from random import uniform
from random import randint

def generateNewValue(lim1, lim2):
    return uniform(lim1, lim2)

def binToInt(x):
    val = 0
    # x.reverse()
    for bit in x:
        val = val * 2 + bit
    return val

def generatePermutation(n):
    p = []
    for i in range(0, n):
        p.append(i)

    random.shuffle(p)

    return p

def generateARandomPermutation(n):
    perm = [i for i in range(n)]
    pos1 = randint(0, n - 1)
    pos2 = randint(0, n - 1)
    perm[pos1], perm[pos2] = perm[pos2], perm[pos1]
    return perm

def generateNedgesWeightedMatrixToFile(no_of_edges, max_weight, output_file_name):
    file_output = 'data/' + output_file_name + '/' + output_file_name + '.txt'
    f = open(file_output, "w")

    f.write(str(no_of_edges))
    f.write('\n')

    mat = []
    for _ in range(no_of_edges):
        l = [] * no_of_edges
        mat.append(l)

    #print(mat)

    for i in range(no_of_edges):
        for j in range(no_of_edges):
            if (i == j):
                mat[i].append(0)
            elif (j > i):
                x = randint(1, max_weight)
                mat[i].append(x)
            else:
                mat[i].append(mat[j][i])

    #print(mat)

    for i in range(no_of_edges):
        for j in range(no_of_edges):
            f.write(str(mat[i][j]))
            f.write(' ')
        f.write('\n')

    f.close()
    print("Scris cu succes în " + file_output)
```

```python
class GA:
    def __init__(self, param=None, problParam=None):
        self.__param = param
        self.__problParam = problParam
        self.__population = []

    @property
    def population(self):
        return self.__population

    def initialisation(self):
        for _ in range(0, self.__param['popSize']):
            c = Chromosome(self.__problParam)
            self.__population.append(c)


    def evaluation(self):
        for c in self.__population:
            c.fitness = self.__problParam['function'](c.repres, self.__problParam['network'], self.__problParam['start'], self.__problParam['end'])

    def bestChromosome(self):
        best = self.__population[0]
        for c in self.__population:
            if (c.fitness < best.fitness):
                best = c
        return best

    def worstChromosome(self):
        best = self.__population[0]
        for c in self.__population:
            if (c.fitness > best.fitness):
                best = c
        return best

    def selection(self):
        pos1 = randint(0, self.__param['popSize'] - 1)
        pos2 = randint(0, self.__param['popSize'] - 1)
        if (self.__population[pos1].fitness < self.__population[pos2].fitness):
            return pos1
        else:
            return pos2

    def oneGeneration(self):
        newPop = []
        for _ in range(self.__param['popSize']):
            p1 = self.__population[self.selection()]
            p2 = self.__population[self.selection()]
            off = p1.crossover(p2)
            off.mutation()
            newPop.append(off)
        self.__population = newPop
        self.evaluation()

    def oneGenerationElitism(self):
        newPop = [self.bestChromosome()]
        for _ in range(self.__param['popSize'] - 1):
            p1 = self.__population[self.selection()]
            p2 = self.__population[self.selection()]
            off = p1.crossover(p2)
            off.mutation()
            newPop.append(off)
        self.__population = newPop
        self.evaluation()

    def oneGenerationSteadyState(self):
        for _ in range(self.__param['popSize']):
            p1 = self.__population[self.selection()]
            p2 = self.__population[self.selection()]
            off = p1.crossover(p2)
            off.mutation()
            off.fitness = self.__problParam['function'](off.repres)
            worst = self.worstChromosome()
            if (off.fitness < worst.fitness):
                worst = off
```

```python
# Generates a given number of new generations after a certain evolutive rule and detects the best ever chromosome
# Input: gaParam, dictionary, parametres of the genetic algorithm, problParam, dict, parametres of the problem,
#        k, number of desired generations
# Output: bestChromo, the best chromosome generated in the k generations

def myGeneticAlgorithm(gaParam, problParam, k):
    k = None

    ga = GA(gaParam, problParam)
    ga.initialisation()
    ga.evaluation()

    bestChromo = ga.bestChromosome()

    firstBestGen = 0

    for g in range(gaParam['noGen']):

        #print('\nGeneration ' + str(g) + '\n')

        # ga.oneGeneration()
        ga.oneGenerationElitism()
        # ga.oneGenerationSteadyState()

        currentBestChromo = ga.bestChromosome()

        if (currentBestChromo.fitness < bestChromo.fitness):
            if (k == None):
                bestChromo = currentBestChromo
                firstBestGen = g

            elif (len(Counter(currentBestChromo.repres).keys()) == k):
                bestChromo = currentBestChromo

        # bestCommunity = ga.bestChromosome().repres
        # bestFitness = ga.bestChromosome().fitness

        #print("Gen.: ", g, "; # comm.: ", len(Counter(currentBestChromo.repres).keys()),  "; Comm.: ", currentBestChromo.repres, '; f(x): ', str(bestChromo.fitness))

    allBestChromo = []

    for x in ga.population:
        if x.fitness == bestChromo.fitness:
            allBestChromo.append(x)

    dict = {'allBestChromo': allBestChromo, 'firstBestGen': firstBestGen}

    return dict
```

```python
# Plots a given graph, its info, its communities generated by my a Genetic algorithm and the communities in a plotted graph
# Input: file_name, name of the file containing a GML-like graph, no_of_communties, integer representing the desired
#        number of communities for the graph
# Output: visual representation of the graph and its communities

def input(file_name, start = None, end = None):
    file_output = 'data/' + file_name + '/' + 'my_' + file_name + '.dat'
    file_path = 'data/' + file_name + '/' + file_name + '.gml'

    # network = nx.read_gml(file_path, label='id')
    network = readNet(file_name)

    MIN = 1
    MAX = nx.number_of_nodes(network)

    noDim = nx.number_of_nodes(network)
    popSize = 300
    noGen = 400

    if (start != None):
        start -= 1
        end -= 1

    # initialise de GA parameters
    gaParam = {'popSize': popSize, 'noGen': noGen}
    # problem parameters
    problParam = {'min': MIN, 'max': MAX, 'function': path_length, 'noDim': noDim, 'noBits': 8, 'network': network, 'start': start, 'end': end}

    bestDict = myGeneticAlgorithm(gaParam, problParam, k=None)

    bestChromo = bestDict['allBestChromo']
    firstGen = bestDict['firstBestGen']

    bestRepres = []

    for x in bestChromo:
        bestRepres.append(x.repres)

    #print(len(bestRepres))
    unique_data = [list(x) for x in set(tuple(x) for x in bestRepres)]

    #print(len(unique_data))
    #uni = set(bestRepres)

    print('\nBest solution after ' + str(noGen) + '\'th generation is', str(len(unique_data)), 'paths: \n')

    unique_d = []

    if (start == None):

        for i in range(0, unique_data.__len__()):
            print("Path", str(i), ":")
            for x in unique_data[i]:
                print(x, "->", end=" ")
            print(unique_data[i][0])
            print('\n')

    else:

        new_set = []

        for i in range(0, unique_data.__len__()):
            lista = []

            #print("Path", str(i), ":")

            start_index = unique_data[i].index(start)
            end_index = unique_data[i].index(end)

            if end_index < start_index:
                start_index, end_index = end_index, start_index

            for x in range(start_index, end_index + 1):
                # print(unique_data[i][x], "->", end=" ")
                lista.append(unique_data[i][x])

            new_set.append(lista)

        unique_d = [list(x) for x in set(tuple(x) for x in new_set)]

        for i in range(0, unique_d.__len__()):
            print("Path", str(i), ":")
            for x in unique_d[i]:
                print(x, "->", end=" ")

            print('\n')

    print('Length:' + str(bestChromo[0].fitness) + '\n')
    print('Found at generation:', firstGen)
```

```python
# Prints the info of a given GML-like graps
# Input: file_name, name of input file
# Output: details about the read graph, its number of nodes and edges

def path_length(path, network, start = None, end = None):
    # noNodes = param['noNodes']
    # mat = param['mat']
    # degrees = param['degrees']
    # noEdges = param['noEdges']
    noNodes = nx.number_of_nodes(network)
    mat = nx.to_numpy_matrix(network)
    degrees = [val for (node, val) in network.degree()]
    noEdges = nx.number_of_edges(network)

    if (start == None):
        new_path = deepcopy(path)
        new_path.append(new_path[0])  # makes path return to first node

        return nx.path_weight(network, new_path, weight='weight')

    else:
        index_start = path.index(start)
        index_end = path.index(end)

        if (index_end < index_start):
            index_start, index_end = index_end, index_start

        new_path = []

        for x in range(index_start, index_end + 1):
            new_path.append(path[x])

        return nx.path_weight(network, new_path, weight='weight')
```

```python
# read the network details
def readNet(file_name):
    file_path = 'data/' + file_name + '/' + file_name + '.txt'

    f = open(file_path, "r")
    net = {}
    n = int(f.readline())
    net['noNodes'] = n
    mat = []
    for i in range(n):
        mat.append([])
        line = f.readline()
        elems = line.split(" ")
        for j in range(n):
            mat[-1].append(int(elems[j]))
    net["mat"] = mat
    degrees = []
    noEdges = 0
    for i in range(n):
        d = 0
        for j in range(n):
            if (mat[i][j] == 1):
                d += 1
            if (j > i):
                noEdges += mat[i][j]
        degrees.append(d)
    net["noEdges"] = noEdges
    net["degrees"] = degrees
    f.close()

    A = np.matrix(net["mat"])
    G = nx.from_numpy_matrix(A)
    # return net
    return G
```
